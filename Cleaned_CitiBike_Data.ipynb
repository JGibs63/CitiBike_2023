```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
import csv
import os
pd.set_option('display.max_rows', None)
import datetime as dt
import glob
```

```python
files = glob.glob('/Users/joshuagibson/Desktop/ClassWork/VU-VIRT-DATA-PT-08-2023-U-LOLC/lesson/18-Tableau/Citi_Bike_Project/Data/*.csv')


# Initialize an empty list to store DataFrames
csv_list = []

# Loop over the files
for file in files:
    df = pd.read_csv(file)
    csv_list.append(df)

# Concatenate DataFrames
combined_df = pd.concat(csv_list, ignore_index=True)

# Display the combined DataFrame
combined_df.head()
```

```python
date_column = pd.to_datetime(combined_df['started_at'])

# Find the minimum and maximum dates
min_date = date_column.min()
max_date = date_column.max()

# Display the results
print("Minimum Date:", min_date)
print("Maximum Date:", max_date)
```
